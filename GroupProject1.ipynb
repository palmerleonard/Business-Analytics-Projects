```python
import numpy as np

df.replace([-1, '-1'], np.nan, inplace=True)
```

```python
#cleaning of job titles

def job_title(title):
  title_lower = title.lower()
  if 'data scientist' in title_lower:
    return 'Data scientist'
  elif 'data engineer' in title_lower:
    return 'Data engineer'
  elif 'analyst' in title_lower:
    return 'Analyst'
  elif 'machine learning' in title_lower:
    return 'MLE'
  elif 'manager' in title_lower:
    return 'Manager'
  elif 'director' in title_lower:
    return 'Director'
  else:
    return 'na'

df['Updated Job Title'] = df['Job Title'].apply(job_title)
```

```python
#List of the most common job types
df['Updated Job Title'].value_counts()
```

```python
# Salary Estimate Columns
sorted(df['Salary Estimate'].unique())
```

```python
#removing 'Employer est.' and 'Glassdoor est.'
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x:x.replace('(Employer est.)','').replace(' (Glassdoor est.)','').replace('$',''))
sorted(df['Salary Estimate'].unique())
```

```python
# Remove letter K, and multiply values with 1000

def salary_estimate(values):
    df = values.replace('K','')
    data2 = df.split('-')
    data3 = [str(int(datas)*1000) for datas in data2]
    return '-'.join(data3)

df['Salary Estimate'] =  df['Salary Estimate'].apply(salary_estimate)

sorted(df['Salary Estimate'].unique())
```

```python
# Adding columna 'min_salary','max_salary', and 'avg_salary'

df['Min Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
df['Max Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
# Calculate average salary
df['Average Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
df['Average Salary'] = df['Average Salary'].astype(int)
```

```python
sorted(df['Company Name'].unique())
```

```python
#Cleaning of company name
df['Company Name'] = df['Company Name'].replace(r'\n\d\.\d', '', regex=True)

sorted(df['Company Name'].unique())
```

```python
print(sorted(df.Rating.unique()))
```

```python
sorted(df['Rating'].unique())

df['Rating'] = np.where(df['Rating'] == -1,0,df['Rating'])
print(sorted(df.Rating.unique()))
```

```python
import pandas as pd
import plotly.express as px

# The dataframe `df` is already loaded and cleaned from previous steps.
# We will use the existing `df` directly.

# Calculate the average salary for each 'Updated Job Title'
avg_salary_by_job = df.groupby('Updated Job Title')['Average Salary'].mean().reset_index()

# Remove the 'na' category from the aggregated data
avg_salary_by_job = avg_salary_by_job[avg_salary_by_job['Updated Job Title'] != 'na']

# Create bar chart
fig = px.bar(
    avg_salary_by_job,
    x="Updated Job Title",
    y="Average Salary",
    color="Updated Job Title",
    labels={
        "Updated Job Title": "Job Title",
        "Average Salary": "Average Salary ($)"
    },
    title="Average Salary by Job Title (Excluding 'na' Category)",
    hover_data=["Updated Job Title", "Average Salary"]
)

# Add maximum value at the top of each bar
for i, row in avg_salary_by_job.iterrows():
    fig.add_annotation(
        x=row["Updated Job Title"],
        y=row["Average Salary"],
        text=f"${row['Average Salary']:.0f}",
        showarrow=False,
        yshift=10
    )

# Improve layout
fig.update_layout(
    legend_title_text="Job Title",
    template="plotly_white"
)

fig.show()
```
